```python
#define x and y variables for model, run both 'Political Lean' and 'Subreddit' for initial NB 

Vectorizer = CountVectorizer(stop_words={'english'})

X = Vectorizer.fit_transform(df_titles['Title'])

y = df_titles['Political Lean']
```

```python
#split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#NaiveBayes - initial model, no stemming, lemmatizing, or use of ngrams

nb_reddit = MultinomialNB()
nb_reddit.fit(X_train,y_train)

nb_reddit_scores = cross_val_score(nb_reddit, X_train, y_train, cv=10)

print(nb_reddit_scores)

print(nb_reddit_scores.mean())
print(nb_reddit_scores.std())
```

With an accuracy score of 32.5%, this is not a great classification model for identifying subreddit categories. Only one in three titles are accurately classified by subreddit. When the y variable is changed to 'Political Lean' the model is able to correctly categorize the title 75.4% of the time. Because the standard deviation is low (.016), the scores are consistent.

*   User location. Because the website Reddit has a global reach, some opinions or topics that are discussed might not be applicable if politicians, campaign managers if they are looking at the local level.
*   Internet trolls (disingenuous users). Many people use sites like Reddit and Twitter with the sole purpose of stirring up controversy and aggravating other users. With this model, there is no difference between someone sharing a genuine opinion and someone seeking negative attention.
*   Political Lean and Subreddit. There is no crossover between subreddit categories and political lean. Someone could have a liberal biased post (based on the language), but since it was posted in a conservative subreddit, it could be falsely classified as conservative. This is most notable, in my mind, with the Libertarian subreddit. In theory, Libertarianism is economically conservative *but* socially liberal. With this dataset, however, any posts that were in the Libertarian subreddit were classified as conservative.

With more time, I would like to explore positive and negative sentiment analysis between the two political biases. Are conservative posters more likely to write positive posts than their liberal counterparts? Does the subreddit matter for this? I would also like to see if user engagement changes between the two groups. Lastly, I would like to see which topics have the most engagement.
